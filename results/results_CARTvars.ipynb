```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_excel("E:/thesis_images/cart_vars_used.xlsx", sheet_name='Blad2', index_col="Unnamed: 0")
df = df.replace(np.nan, 0)
df.loc[:, 'Sum']= df.sum(axis=1)
df = df.sort_values(by='Sum', ascending=False)
```

```python
# Delete rows with only 0s
df = df.loc[(df != 0).any(axis=1), :]
df
```

```python
plt.figure(figsize=(12, 12))
_ = sns.heatmap(df.loc[:, :'Shannons'], cbar=False, cmap='coolwarm', vmin=0, vmax=1, square=True, linewidths=1, linecolor='black')
plt.title('Use of model variables in CART trees for each metric', fontsize=18)
plt.savefig('E:/thesis_images/cart/cart_var_usage.png', format='png', bbox_inches="tight")
```
